```python
import numpy as np
import os

# --- Configure ranges ---
k_values   = [2]   # k-point functions
ell_values = [1, 2, 3]      # loop orders

results = {}  # (k, ell) -> counts dict

for k_ in k_values:
    for ell_ in ell_values:
        print('\n--- k={}, ell={} ---'.format(k_, ell_))
        _, _, _, counts_ = enumerate_all(k_, ell_, verbose=True)
        results[(k_, ell_)] = counts_

# --- Save to npz ---
os.makedirs('counts', exist_ok=True)

k_arr     = [k_   for (k_, ell_) in results]
ell_arr   = [ell_ for (k_, ell_) in results]
trees_arr = [v['n_trees']       for v in results.values()]
topos_arr = [v['n_topologies']  for v in results.values()]
pds_arr   = [v['n_prediagrams'] for v in results.values()]

npz_path = 'counts/enumeration_counts.npz'
np.savez(
    npz_path,
    k=np.array(k_arr),
    ell=np.array(ell_arr),
    n_trees=np.array(trees_arr),
    n_topologies=np.array(topos_arr),
    n_prediagrams=np.array(pds_arr),
)
print('\nSaved to ' + npz_path)

# --- Pretty-print table ---
print('\n' + '{:>4} {:>4} {:>8} {:>8} {:>10}'.format('k', 'ell', 'trees', 'topos', 'prediags'))
print('-' * 38)
for (k_, ell_), v in sorted(results.items()):
    print('{:>4} {:>4} {:>8} {:>8} {:>10}'.format(
        k_, ell_, v['n_trees'], v['n_topologies'], v['n_prediagrams']))
```
