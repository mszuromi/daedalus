```python
# Load Diagram Data from NPZ Files

import numpy as np
import os
from pathlib import Path

print("Diagram Data Loader")
print("="*60)
```

```python
# Cell: Find Available Diagram Sets

# Find all diagram directories
diagram_dirs = [d for d in os.listdir('.') if d.startswith('diagrams_k') and os.path.isdir(d)]

print("Available diagram sets:\n")
for d in sorted(diagram_dirs):
    # Count files in directory
    npz_files = [f for f in os.listdir(d) if f.endswith('.npz')]
    print(f"  {d}: {len(npz_files)} diagrams")

if not diagram_dirs:
    print("  No diagram directories found!")
    print("  Run the enumeration notebook first.")
```

```python
# Cell: Load Diagrams for Specific (k, ell)

# Specify which diagrams to load
k = 2
ell = 1

diagram_dir = f'diagrams_k{k}_ell{ell}'

if not os.path.exists(diagram_dir):
    print(f"Directory {diagram_dir} not found!")
else:
    # Load all diagrams
    npz_files = sorted([f for f in os.listdir(diagram_dir) if f.endswith('.npz')])
    
    diagrams = []
    
    print(f"Loading diagrams from {diagram_dir}/\n")
    
    for npz_file in npz_files:
        filepath = os.path.join(diagram_dir, npz_file)
        data = np.load(filepath, allow_pickle=True)
        diagrams.append(data)
        
        print(f"Loaded: {npz_file}")
    
    print(f"\nTotal diagrams loaded: {len(diagrams)}")
```

```python
# Cell: Inspect a Specific Diagram

# Choose which diagram to inspect
diagram_num = 1  # Change this to inspect different diagrams (1-indexed)

if diagram_num < 1 or diagram_num > len(diagrams):
    print(f"Invalid diagram number! Must be between 1 and {len(diagrams)}")
else:
    data = diagrams[diagram_num - 1]
    
    print(f"{'='*60}")
    print(f"DIAGRAM {diagram_num}")
    print(f"{'='*60}\n")
    
    # Metadata
    print(f"Metadata:")
    print(f"  k (external points): {data['k']}")
    print(f"  ℓ (loop order): {data['ell']}")
    print(f"  Symmetry factor: {data['symmetry_factor']}")
    print(f"  Free parameters: {data['num_free_params']}")
    
    # Graph structure
    print(f"\nGraph Structure:")
    vertices = data['vertices']
    edges = data['edges']
    vertex_types = data['vertex_types']
    
    print(f"  Vertices: {list(vertices)}")
    print(f"  Edges: {len(edges)}")
    
    print(f"\n  Vertex Types:")
    for v, vtype in zip(vertices, vertex_types):
        print(f"    v{v}: {vtype}")
    
    print(f"\n  Edge List:")
    for i, (u, v) in enumerate(edges):
        print(f"    ({u}→{v})")
    
    # Frequencies
    print(f"\nFrequencies:")
    omega_vars = data['omega_variables']
    print(f"  Free parameters: {list(omega_vars)}")
    
    edge_freqs = data['edge_frequencies']
    print(f"\n  Edge Frequencies:")
    for i, (u, v) in enumerate(edges):
        print(f"    ({u}→{v}): {edge_freqs[i]}")
    
    # Times
    print(f"\nTimes:")
    vertex_times = data['vertex_times']
    print(f"  Vertex Times:")
    for v, t in zip(vertices, vertex_times):
        print(f"    v{v}: {t}")
    
    edge_times_from = data['edge_times_from']
    edge_times_to = data['edge_times_to']
    print(f"\n  Edge Times (from, to):")
    for i, (u, v) in enumerate(edges):
        print(f"    ({u}→{v}): ({edge_times_from[i]}, {edge_times_to[i]})")
    
    source_edge_times = data['source_edge_times']
    if len(source_edge_times) > 0:
        print(f"\n  Source Edge Times:")
        for info in source_edge_times:
            print(f"    {info}")
    
    print(f"\n{'='*60}")
```

```python
# Cell: Summary of All Diagrams

print(f"{'='*60}")
print(f"SUMMARY OF ALL DIAGRAMS (k={k}, ℓ={ell})")
print(f"{'='*60}\n")

for i, data in enumerate(diagrams):
    num_vertices = len(data['vertices'])
    num_edges = len(data['edges'])
    num_free = data['num_free_params']
    symmetry = data['symmetry_factor']
    omega_vars = ', '.join(data['omega_variables'])
    
    print(f"Diagram {i+1}:")
    print(f"  Vertices: {num_vertices}, Edges: {num_edges}")
    print(f"  Free params: {num_free} ({omega_vars})")
    print(f"  Symmetry factor: 1/{symmetry}")
    print()

print(f"{'='*60}")
```

```python
# Cell: Example - Extract Specific Information

print("Example: Extract edge frequencies for all diagrams\n")

for i, data in enumerate(diagrams):
    print(f"Diagram {i+1} edge frequencies:")
    edges = data['edges']
    freqs = data['edge_frequencies']
    
    for (u, v), freq in zip(edges, freqs):
        print(f"  ({u}→{v}): {freq}")
    print()
```
